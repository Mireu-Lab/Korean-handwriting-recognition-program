```python
import matplotlib.pyplot as plt
import matplotlib.patches as patches
```

```python
# 1. 기본 스타일 설정
plt.style.use('default')
plt.rcParams['figure.figsize'] = (5, 4)
plt.rcParams['font.size'] = 12

# 2. 사각형 정보를 나타내는 튜플의 리스트
rect = [
    (0, 174, 30, 176), 
    (1821, 143, 1860, 157)
]

# 3. add_patch()로 도형 (patches) 시각화하기
fig, ax = plt.subplots()

for (x, y, w, h) in rect:
    ax.add_patch(
        patches.Rectangle(
            (x, y), w, h,
            edgecolor='green',
            facecolor='palegreen',
            linewidth=0.5,
            fill=True))

# 4. 그래프 설정 및 띄우기
ax.set_xlim(0, 1000)
ax.set_ylim(0, 800)
plt.show()
```
